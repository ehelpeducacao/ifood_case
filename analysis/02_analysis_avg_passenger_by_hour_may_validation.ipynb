# Qual a média de passageiros passenger_count por cada hora do dia que pegaram táxi no mês de maio considerando todos os táxis da frota?


> ### _Resposta usando Pyspark_

```python
from pyspark.sql import functions as F
```

```python
GOLD_HOURLY_TABLE = "ifood_case.gold_hourly_metrics"

df_gold_hourly = spark.table(GOLD_HOURLY_TABLE)

df_avg_passenger_by_hour_may = (
    df_gold_hourly
    .filter(F.col("pickup_month") == "2023-05")
    .groupBy("pickup_hour")
    .agg(
        F.round(
            F.sum("total_passengers") / F.sum("total_trips"),
            2
        ).alias("avg_passenger_count")
    )
    .orderBy("pickup_hour")
)

display(df_avg_passenger_by_hour_may)
```
