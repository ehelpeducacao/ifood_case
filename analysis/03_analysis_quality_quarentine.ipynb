# Analise de qualidade e dados em quarentena.

##  Report Sumarizado

```python
%sql
SELECT 
    total_records,
    valid_records,
    invalid_records,
    valid_percentage,
    invalid_percentage,
    report_timestamp
FROM   workspace.ifood_case.quality_summary_taxi_trips
```

## Registros Rejeitados

```python
%sql
SELECT
    pickup_date,
    taxi_type,
    rejection_reason,
    COUNT(*) AS total_rejections
FROM workspace.ifood_case.quarantine_taxi_trips
WHERE pickup_month BETWEEN '2023-01' AND '2023-05'
GROUP BY
    pickup_date,
    taxi_type,
    rejection_reason
ORDER BY
    pickup_date,
    rejection_reason,
    taxi_type,
    total_rejections;
```

## Registros em Quarentena

```python
%sql
SELECT
    pickup_date,
    pickup_month,
    taxi_type,
    COUNT(*) AS total_rejected_records,
    ROUND(SUM(passenger_count), 2) AS total_rejected_passengers,
    ROUND(SUM(total_amount), 2) AS total_amount_not_computed
FROM workspace.ifood_case.quarantine_taxi_trips
WHERE pickup_month BETWEEN '2023-01' AND '2023-05'
GROUP BY
    pickup_date,
    pickup_month,
    taxi_type
ORDER BY
    pickup_date,
    taxi_type,
    total_amount_not_computed DESC;
```

```python
%sql
SELECT
    VendorID,
    taxi_type,
    source_file,
    pickup_datetime,
    dropoff_datetime,
    pickup_date,
    pickup_hour,
    pickup_month,
    passenger_count,
    total_amount,
    rejection_reason,
    ingestion_timestamp
FROM workspace.ifood_case.quarantine_taxi_trips
WHERE pickup_month BETWEEN '2023-01' AND '2023-05'
ORDER BY
    pickup_date,
    taxi_type,
    rejection_reason,
    pickup_datetime;
```
