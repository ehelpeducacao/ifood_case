# Qual a média de valor total (total\_amount) recebido em um mês considerando todos os yellow táxis da frota?

> ### _Reposta usando Pyspark_

```python
from pyspark.sql import functions as F
```

```python
GOLD_DAILY_TABLE = "ifood_case.gold_daily_metrics"
df_gold = spark.table(GOLD_DAILY_TABLE)
df_avg_total_amount_yellow = (
    df_gold
    .filter(F.col("taxi_type") == "yellow")
    .filter(F.col("pickup_month").between("2023-01", "2023-05"))
    .groupBy("pickup_month")
    .agg(
        F.round(
            F.sum("sum_total_amount") / F.sum("total_trips"),
            2
        ).alias("avg_total_amount")
    )
    .orderBy("pickup_month")
)
```

```python
display(df_avg_total_amount_yellow)
```

> ### _Usando SQL_

```python
%sql
SELECT
    pickup_month,
    ROUND(
        SUM(sum_total_amount) / SUM(total_trips),
        2
    ) AS avg_total_amount
FROM ifood_case.gold_daily_metrics
WHERE taxi_type = 'yellow'
  AND pickup_month BETWEEN '2023-01' AND '2023-05'
GROUP BY pickup_month
ORDER BY pickup_month;
```
